```python
# IMPORT THE LIBRARY AND DATASET INTO NOTEBOOK FOR UNDERSTANDING AND CLEANING.
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('SalesData.csv')
df
```

```python
# LOOKING AT THE FIRST FIVE ROW AND LAST FIVE ROW WE CAN UNDERSTAND THE DATASET HAS ALOT OF INCONSISTENT FORMAT AND MISSING VALUES.
# LETS CHECK THE SHAPE OF DATAFRAME WITH TOTAL COLUMN AND ROWS.
```

```python
df.shape
```

```python
# THE DATAFRAME HAS TOTAL 13 COLUMN WITH MORE THEN 1000 ROWS, NEXT WE NEED TO CHECK MISSING VALUES AND IF THE VALUES ARE IN CORRECT DATATYPE.
```

```python
df.info()
```

```python
# FROM THE GIVEN INFO PRICE IS NOT IN ACCURATE DATATYPE AND 8 COLUMNS HAS MISSING VALUES IN IT.
```

```python
df['Price'] = df['Price'].astype(str)
df['Price'] = df['Price'].str.replace(',', '', regex=False)
df['Price'] = df['Price'].str.replace('MMK', '', regex=False)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df
```

```python
# AFTER CHANGING THE DATATYPE NOW ITS TIME FO DROP AND FILL EMPTY ROWS
```

```python
df.isnull().sum()
```

```python
# DEALING WITH EMPTY ROW SINCE PRICE ONLY HAD 9 MISSING VALUES I DECIDE TO DROP IT RATHER THEN FILL VALUE AND BRING OUTLIERS. 
# FOR CUSTOMER RSTING AND DELIVERY DAY I USED REPLACED WITH MEDIAN VALUE.
# IF THE DISCOUNT ROW IS EMPTY THEN ITS NO DISCOUNT HENCE ZERO PERCENT DISOUNT.
```

```python
df = df.dropna(subset=['Price'])
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())
df['Delivery_Days'] = df['Delivery_Days'].fillna(df['Delivery_Days'].median())
df['Discount_%'] = df['Discount_%'].fillna(0)
df['Sales_Channel'] = df['Sales_Channel'].fillna('Unknown')
df['Region'] = df['Region'].fillna('Unknown')
df
```

```python
# NOW GENERATE NEW FEATURE FROM ALREADY EXISTING COLUMN TO CREATE MORE INSIGHTFUL MESSAGE.
```

```python
df['Discount'] = df['Quantity'] * df['Price'] * (df['Discount_%'] / 100)
df['Amount'] = df['Quantity'] * df['Price'] * (1 - df['Discount_%'] / 100)
df
```

```python
# CREATING NEW COLUMN NOW ITS TIME FOR DROPPING ANY VALUELESS ATTRIBUTE
```

```python
df = df.drop(columns=[
    'Date',
    'Discount_%',
    'Payment_Method',
    'Customer_Type',
    'Delivery_Days',
    'Price',
    'Quantity',
    'Product',
    'Customer_Rating'
])
df = df.dropna(subset=['Total_Sales'])
df
```

```python
# NOW THAT I HAVE DROP ANY USELESS COLUMN AND EMPTY ROWS LETS CHECK DATATYPE AND NULL VALUES AGAIN
```
